## This is a sample pandas project that exemplifies how to implement web scraping and crawler, specifically for data science and machine learning purposes, using Python, Pandas, time, requests and BeautifulSoup, along with Anaconda managed environments

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
# At this project, we use a sample book store URL to retrieve the books info from the first 10 pages of 
# its catalogue - by using web scraping and a simple web crawler
```

```python
URL = "https://books.toscrape.com/"
res = requests.get(URL)
res.encoding = "utf-8"
print("status_code:", res.status_code)
soup = BeautifulSoup(res.text)

# get the URL requests response and pass it to BeautifulSoup instance
```

```python
books = soup.find_all("article", class_="product_pod")
print("Books on this page:", len(books))

data = []
for book in books:
    title = book.find("h3").find("a")["title"]
    price  = book.find("p", class_="price_color").get_text(strip=True)
    rating   = book.find("p", class_="star-rating")["class"][1]
    stock = book.find("p", class_="instock").get_text(strip=True)

    data.append({"title": title, "price": price, "rating": rating, "stock": stock})
    print(f"{rating:6} | {price:8} | {title}")

# use the BeautifulSoup instance to navigate the HTML structure, either getting elements with find or 
# find_all, filtering by class - including by the second class of certain elements (rating), then get
# the text and show each book info
```

```python
for book in soup.select("article.product_pod h3 a"):
    print("-", book["title"])

# you can also use select at BeautifulSoup, instead of find / find_all
# selecting only the titles, in this case
```

```python
# Using the pattern of the catalogue pages to exemplify a simple crawler that goes through all the first
# 10 pages and brings books info as a return
# URL pattern:
# https://books.toscrape.com/catalogue/page-N.html

all_books = []
for page in range(1, 11):
    url = f"https://books.toscrape.com/catalogue/page-{page}.html"
    r = requests.get(url)
    r.encoding = "utf-8"
    if r.status_code != 200:
        break
    s = BeautifulSoup(r.text, "lxml")
    for book in s.find_all("article", class_="product_pod"):
        all_books.append({
            "title": book.find("h3").find("a")["title"],
            "price": book.find("p", class_="price_color").get_text(strip=True),
            "rating": book.find("p", class_="star-rating")["class"][1],
        })
    print(f"Page {page}: total = {len(all_books)} livros")
    time.sleep(1)
```

```python
df = pd.DataFrame(all_books)
df.to_csv("books.csv", index=False)
print("Total books:", len(df))
df
```
