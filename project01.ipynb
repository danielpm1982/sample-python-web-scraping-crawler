## This is a sample pandas project that exemplifies how to implement web scraping and crawler, specifically for data science and machine learning purposes, using Python, Pandas, time, requests and BeautifulSoup, along with Anaconda managed environments

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# At this project, we get a sample static HTML page content, using the requests lib, then we add
# that string content to an instance of BeautifulSoup lib, in order to select individually each element 
# we want, for each class of that element, and then get the text of the element to mount a dataframe and 
# save that to a .csv file. This way we extract all data from the elements of this sample HTML page and
# structure the data into a dataframe and csv file.
```

```python
URL = "https://guilhermeonrails.github.io/clima-tempo/"

# asking HTML content of the URL page
res = requests.get(URL)

print("status_code:", res.status_code)
print("Characters at res.text:", len(res.text))

print("\n----- HTML CRU (initial) -----")
print(res.text[:1000])
```

```python
# loads res.text (HTML) at the BeautifulSoup object
soup = BeautifulSoup(res.text, "lxml")

# find: finds the first HTML element. In this case, the first <h1> .
title = soup.find("h1")
print("Page title:", title.get_text(strip=True))
```

```python
# find_all: finds all div cards with bg-white class (all cities)
cards = soup.find_all("div", class_="bg-white")
print("Found cities:", len(cards))
```

```python
# For each city card, we extract the three respective variables: temperature, weather and humidity
# see full HTML structure at the URL page above
data = []
print(f"{'CITY':15} | {'TEMP':6} | {'WEATHER':22} | {'HUMIDITY'}")
print(f"{'==============':15} | {'====':6} | {'====================':22} | {'==========='}")
for card in cards:
    city = card.find("h2").get_text(strip=True)
    values = card.find_all("span", class_="font-bold")
    temperature = values[0].get_text(strip=True)
    weather       = values[1].get_text(strip=True)
    humidity     = values[2].get_text(strip=True)
    data.append({
        "city": city,
        "temperature": temperature,
        "weather": weather,
        "humidity": humidity,
    })
    print(f"{city:15} | {temperature:6} | {weather:22} | umidade {humidity}")
```

```python
# adding the dictionary list (each city dictionary) to a Pandas dataframe
df = pd.DataFrame(data)
df
```

```python
# saving df data to a local .csv file
df.to_csv("weather.csv", index=False)
print("Saved !")
```
